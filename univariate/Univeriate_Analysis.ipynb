```python
import pandas as pd
dataset=pd.read_csv("Placement.csv")
dataset
```

```python
dataset=dataset.drop(columns=["sl_no"])
dataset
```

```python

from univariate import univariate
quan,qual=univariate.quanqual(dataset)
```

```python
quan
```

```python
qual
```

```python
#Descriptive with outlier
descriptive=univariate.descriptive(dataset,quan)
descriptive
```

```python
#Finding outliers and Replacing Outliers
lower,upper=univariate.findingoutliers(quan,descriptive)
if not lower or not upper:
    print("There is no potential outliers")
else:
    print("There are potential outliers \n Lower_Outliers=",lower,"\n Upper_Outliers=",upper)
    dataset=univariate.replacingoutliers(lower,upper,descriptive,dataset)
    print("Outliers are Replaced")
    descriptive=univariate.descriptive(dataset,quan)
    
   
```

```python
descriptive
```

```python
for columnName in quan:
    print(columnName,"Frequency Table")
    FreqTable=univariate.frequency(columnName,dataset)
    print(FreqTable)
```
