```python
import pandas as pd
dataset=pd.read_csv("Kidney_disease.csv")
dataset
```

```python
dataset=dataset.drop(columns=["id"])
dataset
```

```python

from univariate import univariate
quan,qual=univariate.quanqual(dataset)
```

```python
quan
```

```python
qual
```

```python
dataset.isnull().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
impquan=SimpleImputer(missing_values=np.nan,strategy='mean')
impquan.fit(dataset[quan])
dataset[quan]=impquan.transform(dataset[quan])
impqual=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
impqual.fit(dataset[qual])
dataset[qual]=impqual.transform(dataset[qual])
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
#Descriptive with outlier
descriptive=univariate.descriptive(dataset,quan)
descriptive
```

```python
#Finding outliers and Replacing Outliers
lower,upper=univariate.findingoutliers(quan,descriptive)
if not lower or not upper:
    print("There is no potential outliers")
else:
    print("There are potential outliers \n Lower_Outliers=",lower,"\n Upper_Outliers=",upper)
    dataset=univariate.replacingoutliers(lower,upper,descriptive,dataset)
    print("Outliers are Replaced")
    descriptive=univariate.descriptive(dataset,quan)
    
   
```
