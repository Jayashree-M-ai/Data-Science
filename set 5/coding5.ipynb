```python
Question 1:

A company wants to analyze whether there is a significant difference in the average sales across four regions using One-Way ANOVA.

Sample Dataset:

    'Region': ['North', 'North', 'South', 'South', 'East', 'East', 'West', 'West'],
    'Sales': [200, 220, 250, 240, 180, 210, 230, 220]

Expected Output:

F-Statistic: 4.866666666666666
P-Value: 0.08022409100499783
```

```python
import pandas as pd
import scipy.stats as st
df=pd.DataFrame({'Region': ['North', 'North', 'South', 'South', 'East', 'East', 'West', 'West'],
    'Sales': [200, 220, 250, 240, 180, 210, 230, 220]})
n=df.loc[df['Region']=='North','Sales']
s=df.loc[df['Region']=='South','Sales']
e=df.loc[df['Region']=='East','Sales']
w=df.loc[df['Region']=='West','Sales']
f,p=st.f_oneway(n,s,e,w)
print("F-Statistic: ",f,"\nP-Value: ",p)
```

```python
Analyze whether the employee performance scores are affected by the type of training and experience using Two-Way ANOVA.

Sample Dataset:

    'Training': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B'],
    'Experience': ['Low', 'High', 'Low', 'High', 'Low', 'High', 'Low', 'High'],
    'Performance': [75, 80, 85, 90, 78, 88, 82, 89]
```

```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
df=pd.DataFrame({'Training': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B'],
    'Experience': ['Low', 'High', 'Low', 'High', 'Low', 'High', 'Low', 'High'],
    'Performance': [75, 80, 85, 90, 78, 88, 82, 89]})
cl=ols('Performance ~ C(Training) + C(Experience) + C(Training):C(Experience)',data=df).fit()
an=anova_lm(cl,type=2)
print(an)
```

```python
Question 3:

A real estate company wants to check for multicollinearity using VIF to predict house prices.

Sample Dataset:

    'Area': [1500, 2000, 2500, 3000, 3500],
    'Bedrooms': [3, 4, 3, 5, 4],
    'Bathrooms': [2, 3, 2, 4, 3],
    'House_Price': [300000, 400000, 350000, 500000, 450000]
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
df=pd.DataFrame({'Area': [1500, 2000, 2500, 3000, 3500],
    'Bedrooms': [3, 4, 3, 5, 4],
    'Bathrooms': [2, 3, 2, 4, 3],
    'House_Price': [300000, 400000, 350000, 500000, 450000]})
x=df.drop(columns='House_Price')
vif=pd.DataFrame()
vif['Features']=x.columns
vif['VIF']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
print(vif)
```

```python

Question 4:

A school tested two teaching methods (Traditional vs Interactive). Perform an Independent T-Test to see if their average scores are different.

Sample Dataset:

    'Method': ['Traditional', 'Traditional', 'Interactive', 'Interactive'],
    'Score': [70, 75, 85, 90]

Expected Output:

Statistic: -4.242640687119285
P-Value: 0.05131670194948621
```

```python
import pandas as pd
from scipy.stats import ttest_ind
df=pd.DataFrame({'Method': ['Traditional', 'Traditional', 'Interactive', 'Interactive'],
    'Score': [70, 75, 85, 90]})
t=df.loc[df['Method']=='Traditional','Score']
i=df.loc[df['Method']=='Interactive','Score']
s,p=ttest_ind(t,i)
print("Statistic: ",s,"\nP-Value: ",p)
```

```python
Question 5:

A fitness center wants to analyze if workout time is positively correlated with calorie burn using Pearson correlation.

Sample Dataset:

data = pd.DataFrame({
    'Workout_Time': [30, 45, 60, 90, 120],
    'Calories_Burned': [200, 300, 400, 500, 600]
})
Expected Output:

Correlation Coefficient: 0.9847982464479192
P-Value: 0.002244820403552132
```

```python
import pandas as pd
from scipy.stats import pearsonr
data = pd.DataFrame({
    'Workout_Time': [30, 45, 60, 90, 120],
    'Calories_Burned': [200, 300, 400, 500, 600]
})
cor,pv=pearsonr(data['Workout_Time'],data['Calories_Burned'])
print("Correlation Coefficient: ",cor,"\nP-Value: ",pv)
```
