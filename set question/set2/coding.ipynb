```python
Question 1 :
An airline tracks flight delays (in minutes) for 20 flights. Analyze the flight delays to calculate percentiles, detect outliers, and evaluate the overall distribution.
DataSet:
delays = [15, 30, 45, 20, 25, 100, 5, 60, 35, 50, 
          120, 80, 10, 75, 90, 200, 55, 40, 70, 150]
Expected Output:

Percentiles (10th, 25th, 50th, 75th, 90th): [ 14.5   28.75  52.5   82.5  123.  ]
IQR: 53.75
Outliers: [200]
```

```python
import numpy as np
import seaborn as sb
delays = [15, 30, 45, 20, 25, 100, 5, 60, 35, 50, 
          120, 80, 10, 75, 90, 200, 55, 40, 70, 150]
p=np.percentile(delays,[10,25,50,75,90])
print("Percentiles (10th, 25th, 50th, 75th, 90th):",p)
IQR=p[3]-p[1]
print("IQR: ",IQR)
lb=p[1]-(1.5*IQR)
ub=p[3]+(1.5*IQR)
outliers=[x for x in delays if(x<lb or x>ub)]
print("Outliers: ",outliers)
sb.boxplot(data=delays, orient='h')
```

```python
Question 2 :
A company wants to analyze the salary distribution of its employees to understand the central tendency and determine whether the data is skewed.
DataSet:
salaries = [30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000,
            60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000]
Expected Output:
Mean Salary: 56950.0
Median Salary: 57500.0
Mode Salary: 30000
The data is Left Skewed (Negative Skew)
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
salaries = [30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000,
            60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000]
mean_val = np.mean(salaries)
median_val = np.median(salaries)
mode_val,count = stats.mode(salaries)
print("Mean Salary:", mean_val)
print("Median Salary:", median_val)
print("Mode Salary:", mode_val)
print("The data is Left Skewed (Negative Skew=>Mean < Median)")
sns.histplot(salaries,kde=True)
plt.title("Histogram of Employee salaries")
plt.xlabel("Salaries")
plt.ylabel("Count")
plt.show()
```

```python
Question 3:
 A school wants to analyze the exam performance of students across three subjects: Mathematics, Science, and English. How can Data Science concepts be applied to understand their performance?
DataSet:
data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Mathematics': [85, 78, 92, 88, 70, 95, 60, 80, 90, 76],
    'Science': [80, 85, 88, 70, 75, 92, 55, 82, 89, 78],
    'English': [78, 74, 85, 80, 68, 90, 50, 77, 83, 72]
}
Expected Output:
Descriptive Statistics
Histogram(graph)
Correlation Analysis(graph)
HeatMap(graph)
```

```python
import pandas as pd
import seaborn as sns
import plotly.express as px
data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Mathematics': [85, 78, 92, 88, 70, 95, 60, 80, 90, 76],
    'Science': [80, 85, 88, 70, 75, 92, 55, 82, 89, 78],
    'English': [78, 74, 85, 80, 68, 90, 50, 77, 83, 72]
}
dataset=pd.DataFrame(data)
des=dataset.describe()
print("Descriptive Statistics:\n", des)
fig=px.histogram(des,title='Histogram of Descriptive Statistics')
fig.show()
sns.heatmap(des,annot=True)
plt.title("Heatmap of Decriptive Statistics")
plt.show()
corr_data=dataset.drop(columns=['Student'])
print("Correlation Analysis:\n",corr_data)
fig=px.histogram(corr_data,title='Histogram of Correlation Analysis')
fig.show()
sns.heatmap(corr_data,annot=True)
plt.title("Heatmap of Correlation")
plt.show()
```

```python
Question 4:
A pharmaceutical company conducted a clinical trial with two groups: one receiving medication and the other a placebo. How do you perform a hypothesis test to determine the effectiveness of the medication?
Dataset:
medication_group = [110, 115, 108, 102, 107, 99, 111, 104, 109, 101]
placebo_group = [120, 125, 130, 122, 128, 119, 124, 127, 123, 126]
Expected Output:
T-Statistic: -9.201427649220966
P-Value: 3.163912817600812e-08
Reject the null hypothesis: The medication is effective.
```

```python
import pandas as pd
from scipy.stats import ttest_ind
data = {'medication_group': [110, 115, 108, 102, 107, 99, 111, 104, 109, 101],
        'placebo_group' : [120, 125, 130, 122, 128, 119, 124, 127, 123, 126]}
df=pd.DataFrame(data)
tt,pv=ttest_ind(df['medication_group'],df['placebo_group'])
print("T-Statistic: ",tt)
print("P-Value:",pv)
if(pv<0.05):
    print("Reject the Null Hypothesis: The medicaion is effective.")
else:
    print("Accept the Null Hypothesis: The medicaion is not effective.")
```

```python
Question 5 : A company conducted a customer satisfaction survey where customers rated their experience on a scale of 1 to 10. Analyze the survey results to calculate descriptive statistics and visualize the distribution of customer satisfaction ratings.
Sample DataSet:  
ratings = [8, 9, 7, 5, 6, 10, 9, 4, 7, 8, 
           6, 9, 10, 5, 8, 7, 6, 9, 10, 7]
Expected Output:
Mean Rating: 7.5
Median Rating: 7.5
Mode Rating: 7
Standard Deviation: 1.746424919657298
```

```python
import numpy as np
ratings = [8, 9, 7, 5, 6, 10, 9, 4, 7, 8, 
           6, 9, 10, 5, 8, 7, 6, 9, 10, 7]
mean_val = np.mean(ratings)
median_val = np.median(ratings)
mode_val,count = stats.mode(ratings)
std=np.std(ratings)
print("Mean Rating: ", mean_val)
print("Median Rating:", median_val)
print("Mode Rating:", mode_val)
print("Standard Deviation:", std)
sns.histplot(ratings,kde=True)
plt.title("Distribution of Customer Satisfaction Ratings")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.show()
```
