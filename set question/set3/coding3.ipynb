```python
Question 1 :
A company wants to analyze customer satisfaction survey data. How can Data Science concepts help to clean and analyze this data?
SampleDataset:
    'CustomerID': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'Satisfaction': [8, 7, np.nan, 9, 6, 6, 10, 4, 5, 7]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.DataFrame({'CustomerID': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'Satisfaction': [8, 7, np.nan, 9, 6, 6, 10, 4, 5, 7]})
data['Satisfaction']=data['Satisfaction'].fillna(data['Satisfaction'].mean())
print(data)
print("Average Satisfaction:", data['Satisfaction'].mean())
print("Median Satisfaction:", data['Satisfaction'].median())
plt.hist(data['Satisfaction'])
plt.title("Customer Satisfaction Distribution")
plt.xlabel("Satisfaction Score")
plt.ylabel("Frequency")
plt.show()
```

```python
Question 2 :
 A retail store has recorded monthly sales data for its products. How can Data Science concepts be applied to understand sales performance?
SampleDataset:
'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
 'Sales': [1500, 1700, 1600, 1800, 2200, 2100]

```

```python
import pandas as pd
import matplotlib.pyplot as plt
sales_data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [1500, 1700, 1600, 1800, 2200, 2100]})
print("Total Sales:", sales_data['Sales'].sum())
print("Average Monthly Sales:", sales_data['Sales'].mean())
plt.plot(sales_data['Month'], sales_data['Sales'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
Question 3:
A hospital conducted a study to check the effectiveness of a new medical test. How can Data Science concepts be used to validate its results?
SampleDataset:
group_a = [110, 115, 108, 102, 107]
group_b = [120, 125, 130, 122, 128]
Expected Output:
T-Statistic: -5.92, P-Value: 0.0004
```

```python
from scipy.stats import ttest_ind
group_a = [110, 115, 108, 102, 107]
group_b = [120, 125, 130, 122, 128]
tt,pv=ttest_ind(group_a,group_b)
print("T-Statistic: ",round(tt,2), "P-Value: ",round(pv,4))
```

```python
Question 4:
 A company wants to evaluate employee performance using various metrics. How can Data Science concepts be applied?
Sample Dataset:
    'Department': ['HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', np.nan, 'HR', 'Sales', 'IT'],
    'Experience (Years)': [1.0, 3.0, 5.0, np.nan, 10.0, 2.0, 8.0, 4.0, np.nan, 9.0],
    'Projects Completed': [2.0, 5.0, np.nan, 10.0, 12.0, 3.0, 9.0, np.nan, 14.0, 11.0],
    'Performance Rating': [3.0, np.nan, 7.0, 8.0, 9.0, 4.0, 8.0, 7.0, 10.0, np.nan]

Perform:
1)	Data Cleaning
2)	Descriptive Statistics
3)	Correlation Analysis (Using heatmap)
4)	Distribution Analysis (Plot Histogram)
5)	Performance Comparison (Using Boxplot)
6)	Experience vs Performance (Perform Bivariate Analysis using a scatter plot)
7)	Department-wise Analysis
```

```python
#import
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot  as plt
df=pd.DataFrame({'Department': ['HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', np.nan, 'HR', 'Sales', 'IT'],
    'Experience (Years)': [1.0, 3.0, 5.0, np.nan, 10.0, 2.0, 8.0, 4.0, np.nan, 9.0],
    'Projects Completed': [2.0, 5.0, np.nan, 10.0, 12.0, 3.0, 9.0, np.nan, 14.0, 11.0],
    'Performance Rating': [3.0, np.nan, 7.0, 8.0, 9.0, 4.0, 8.0, 7.0, 10.0, np.nan]})
df
```

```python
#1)	Data Cleaning
df['Department']=df['Department'].fillna(df['Department'].mode()[0])
quan=['Experience (Years)','Projects Completed','Performance Rating']
df[quan]=df[quan].fillna(df[quan].mean())
df
```

```python
#2)	Descriptive Statistics
print(df.describe())
```

```python
#3)	Correlation Analysis (Using heatmap)
corr=df[quan].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Analysis (Using heatmap)")
plt.show()
```

```python
#4)	Distribution Analysis (Plot Histogram)
for col in quan:
    plt.figure()
    plt.hist(df[col])
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
#5)	Performance Comparison (Using Boxplot)
sns.boxplot(x='Department', y='Performance Rating', data=df)
plt.title("Performance Comparison Across Departments")
plt.show()
```

```python
#6)	Experience vs Performance (Perform Bivariate Analysis using a scatter plot)
sns.scatterplot(x='Experience (Years)',y='Performance Rating',data=df)
plt.title("Experience vs Performance")
plt.xlabel("Experience (Years)")
plt.ylabel("Performance Rating")
plt.show()
```

```python
#7)	Department-wise Analysis
department_summary = df.groupby('Department')[['Experience (Years)', 'Projects Completed', 'Performance Rating']].mean()
print(department_summary)
department_summary.plot(kind='bar')
plt.title("Department wise Analysis")
plt.ylabel("Average value")
plt.show()
```
