```python
import pandas as pd
dataset=pd.read_csv("preproplacement.csv")
dataset
```

Independent Sample - Unpaired T-Test



Different group(male,female) but same condition(salary)

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
male=dataset[dataset['gender']=='M']['salary']
female=dataset[dataset['gender']=='F']['salary']
ttest_ind(male,female)
```

Dependent Sample - Paired T-Test

Same group(male,female) but different condition(salary)

```python
from scipy.stats import ttest_rel
male=dataset[dataset['gender']=='M']['ssc_p']
male1=dataset[dataset['gender']=='M']['hsc_p']
ttest_rel(male,male1)
```

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
o=dataset[dataset['hsc_b']=='Others']['salary']
c=dataset[dataset['hsc_b']=='Central']['salary']
ttest_ind(o,c)
```

```python
from scipy.stats import ttest_rel
od=dataset[dataset['hsc_b']=='Others']['degree_p']
om=dataset[dataset['hsc_b']=='Others']['mba_p']
ttest_rel(od,om)
```

Hypothesis Testing
Statement: Test whether ssc_p and hsc_p of male is same
Null hypothesis - There is no similarity between 10th pass and 12th pass
Alternate hypothesis - There is similarity between 10th pass and 12th pass

```python
from scipy.stats import ttest_rel
male=dataset[dataset['gender']=='M']['ssc_p']
male1=dataset[dataset['gender']=='M']['hsc_p']
ttest_rel(male,male1)
```

pvalue>0.05, so accept Null hypothesis and reject alternate hypothesis

ANAVO

```python
import scipy.stats as stats
male=dataset[dataset['gender']=='M']['salary']
female=dataset[dataset['gender']=='F']['salary']
stats.f_oneway(male,female)
```

Null hypothesis - There is no significant difference between 10th pass,12th pass and degree pass
Alternate hypothesis - There is significant difference between 10th pass,12th pass and degree pass

pvalue>0.05, so accept null hypothesis and reject alternate hypothesis

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model=ols('salary~C(gender)*C(specialisation)',data=dataset).fit()
anova_table=sm.stats.anova_lm(model, typ=2)
anova_table
```
